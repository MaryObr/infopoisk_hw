# Общая функция для использования всех способов поиска

```python
import import_ipynb
from preprocessing import *
from first_search import *
from dict_search import *
from matrix_search import *
```

```python
def general_search(query:str, index_type:str='bm', type_search:str='matrix', top_n:int=5, k=1.5, b=0.75):
    '''
    Принимает запрос, тип индекса (bm25/частотный), тип поиска (базовый, через словари или через матрицы), нужную длину топа документов. Возвращает топ документов по убыванию актуальности

    query:str -- запрос на русском языке
    
    index_type:str -- вид индекса:
        bm25 -- bm25
        frequency -- частотный
        
    type_search:str -- вид подсчета:
        basic -- базовый
        dictonary -- через словари
        matrix -- через матрицы
        
    top_n:int -- длина топа возвращаемых документов
    '''
    if index_type == 'bm':
        if type_search == 'basic':
            return search(query, top_k=top_n)
        elif type_search == 'dictionary':
            return dict_bm25(query, k, b, top_k=top_n) 
        elif  type_search == 'matrix':
            return matrix_bm(query, top_k=top_n, k=k, b=b)
        else:
            raise ValueError(f'Неверный type search -- {type_search}, доступные значения: basic, dictionary, matrix')
            
    elif index_type == 'frequency':
        if type_search == 'basic':
            return search(query, index_type='freq', top_k=top_n)
        elif type_search == 'dictionary':
            return dict_freq(query, top_k=top_n) 
            return matrix_freq(query, top_k=top_n)
        else:
            raise ValueError(f'Неверный type search -- {type_search}, доступные значения: basic, dictionary, matrix')
    else:
        raise ValueError(f'Неверный index type -- {index_type}, доступные значения: bm25, frequency')            
```
